```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
petal_length = df.petal_length
```

```python
petal_width = df.petal_width
```

```python
plt.scatter(x = petal_length, y = petal_width )
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.show()
```

```python
setosa =  df[df.species == "setosa"]
versicolor =  df[df.species == "versicolor"]
virginica =  df[df.species == "virginica"]
```

```python
df.species.unique()
```

```python
plt.scatter(setosa['petal_length'], setosa.petal_width, label = 'setosa', color = 'green')
plt.scatter(versicolor['petal_length'], versicolor.petal_width, label = 'versicolor', color = 'red')
plt.scatter(virginica['petal_length'], virginica.petal_width, label = 'virginica', color = 'yellow')
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.legend()
plt.show()
```

```python
plt.hist(df.petal_length, bins=25)
plt.xlabel("petal length")
plt.ylabel("frequency")
plt.show()
```

```python
df.species.value_counts()
```

```python
plt.pie(x =df.species.value_counts(), labels=df.species.value_counts().index, autopct="%0.1f%%" )
plt.show()
```

```python
movie = pd.read_csv('https://raw.githubusercontent.com/coding-minutes/python-data-science-mastercourse/main/06%20Matplotlib/movie_meta.csv')
```

```python
movie.head()
```

```python
movie.columns
```

```python
movie.movie_title[30]
```

```python
def preprocessing(x):
  return len(x.strip())
```

```python
movie.movie_title.apply(preprocessing)
```

```python
movie['movie_title_len'] = movie.movie_title.apply(lambda x:len(x.strip()))
```

```python
movie.movie_title_len
```

```python
movie.columns
```

```python
plt.hist(movie.movie_title_len, bins=50)
plt.xticks(range(0,100,5))
plt.xlabel("Movie Title length")
plt.ylabel("Number of movies(freq)")
plt.show()
```

```python
d = {}

for i in movie.movie_title_len:
  if i not in d:
    d[i] =  1
  else:
    d[i]+=1
```

```python

d
```

```python
plt.scatter(x = d.keys(), y = d.values())
plt.xticks(range(0,100, 5))
plt.xlabel("Movie title len")
plt.ylabel("Number of movies (Freq)")
plt.show()
```
