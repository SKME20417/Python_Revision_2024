```python
import seaborn as sns
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
import warnings
warnings.filterwarnings('ignore')

sns.distplot(df.petal_length, bins=20, kde = False)
plt.show()
```

```python
sns.kdeplot(df.petal_length)
plt.show()
```

```python
sns.scatterplot(x = df.petal_length, y = df.petal_width)
plt.show()
```

```python
sns.scatterplot(x = 'petal_length', y = 'petal_width', data=df, hue='species')
plt.show()
```

```python
sns.jointplot(x = 'sepal_length', y = 'sepal_width', data=df, kind='reg')
plt.show()
```

```python
sns.pairplot(data = df, hue='species')
```

```python
sns.countplot(x = 'species', data=df)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(x= 'species', y = 'sepal_length', data=df)
plt.show()
```

```python
sns.violinplot(x = 'species', y = 'petal_length', data=df)
plt.show()
```

```python
new_df = df.drop('species', axis=1)
```

```python
new_df.corr()
```

```python
sns.scatterplot(x = 'sepal_length', y = 'sepal_length', data=df)
plt.show()
```

```python
sns.heatmap(new_df.corr(), annot=True, cmap='Blues')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.shape
```

```python
tips.head()
```

```python
sns.countplot(x = 'sex', data=tips)
plt.show()
```

```python
tips.day.unique()
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips, hue='sex', estimator=np.median )
plt.show()
```

```python
sns.barplot(x = 'time', y = 'total_bill', data=tips, hue='day')
plt.show()
```

```python
sns.barplot(x = 'smoker', y = 'tip', data=tips)
plt.show()
```

```python
sns.lmplot(x = 'total_bill', y = "tip", data=tips)
plt.show()
```
