```python
movie = pd.read_csv('https://raw.githubusercontent.com/coding-minutes/python-data-science-mastercourse/main/06%20Matplotlib/movie_meta.csv')
```

```python
movie.head()
```

```python
movie.shape
```

```python
movie.columns
```

```python
movie.movie_title[5]
```

```python
movie['movie_length'] = movie.movie_title.apply(lambda x:len(x.strip()))
```

```python
plt.hist(movie.movie_length, bins=50)
plt.xlabel("Movie Title Length")
plt.ylabel("Number of Movies (freq)")
plt.xticks(range(0,100,5))
plt.show()
```

```python
d = {}
for i in movie.movie_length:
  if i not in d:
    d[i] = 1
  else:
    d[i]+=1
```

```python
d
```

```python
plt.scatter(x = d.keys(), y = d.values())
plt.xlabel("Movie Title Length")
plt.ylabel("Number of Movies (freq)")
plt.xticks(range(0,100,5))
plt.show()
```
